```python
#converting release time to datetime as they can't be object the datetime is always in strings
#This argument is crucial. If there are any values in 'release_date' that cannot be parsed into a datetime (e.g., malformed strings), errors='coerce' will turn them into NaT (Not a Time, pandas' version of NaN for datetimes), preventing errors and allowing the rest of the operation to proceed. NaT values will then correctly result in NaN in the 'release_year' column.
movies['release_date']=pd.to_datetime(movies['release_date'],errors='coerce')
movies['release_year']=movies['release_date'].dt.year
movies['release_year']
```

```python
movies=movies[['genres','id','keywords','title','overview','cast','crew','release_year','popularity']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def extract(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(extract)
```

```python
movies['keywords']=movies['keywords'].apply(extract)
```

```python
movies['cast'][0]
```

```python
def extract_top_3(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=3:
      L.append(i['name'])
      count+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(extract_top_3)
movies['cast'][0]
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#it is a string and we will convert it to list for concatination with others
# Check if the element is not already a list before splitting
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else x)
```

```python
# Remove spaces within column values to prevent inconsistencies in the recommender system
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#merging the columns to create a tag and converting them to string again to create a detailed paragraph for the mvoie tag
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tags'][0]
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#stemming the data to reduce words ex ['loving','love'] to ['love','love']
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Now our work is to recommend 5 similar movies to the user according to the movie or content he's watching
#To identitfy similarity we have to calculate the similarity between the tags of the movies
#therefore we will vectorise the text for that purpose and the techniques used is bag of words which is we will combine all the tags of all the mvoies
#then we will find the top 5000 words that has the most frequency from the concatenated tags
#for all the movies we will compare the rows that will have the frequencies of the top 5000 words for particular movies
#using CountVectorizer library for converting strings to token
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#toarray is used to convert the sparse matrix(from transform) to array
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
#gives us the most frequent words
cv.get_feature_names_out()
```

```python
#calculating cosine distances for finding the similarity between movies
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df[new_df['title']=='Avatar'].index
```

```python
#through enumerate function we could retain the original indexes of the movies even when they are reversed
#reversing the array to find the mostly similary movies on the top and then fetching top 5 similar movies from it
#using lambda function to check for the second value of tuple x (index,distance)
#1:6 is used to extarct top similar movies starting from 1st position as 0th position will have value 1 the highest as the movie have the distance with itself
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  #find the indexes of the movies
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
