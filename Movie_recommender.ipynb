```python
#Removing columns that are not necessary that is preprocessing the data and only including columns that are involved in making tags
movies=movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
#we just want the name key values from the dictionary in form of a list
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
#before using convert function we first need to convert the genres from strings to list using ast module
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
#for cast same thing to be followed but we will just use top five cast
def converts(obj):
  if isinstance(obj, str):  # Check if the object is a string
      l = []
      count = 0
      for i in ast.literal_eval(obj):
          if count != 5:
              l.append(i['name'])
              count += 1
          else:
              break
      return l
  else: # if its not a string return as is.
    return obj
```

```python
def converted(obj):
  if isinstance(obj, str):  # Check if obj is a string before using ast.literal_eval
    l = []
    for i in ast.literal_eval(obj):
      l.append(i['name'])
    return l
  else:  # If obj is not a string (e.g., a list), likely already processed
    # Return as is if it's already a list of strings
    if isinstance(obj, list) and all(isinstance(item, str) for item in obj):
      return obj
    # Handle other cases as needed
    return []  # or raise an exception if this scenario is unexpected
movies['keywords']=movies['keywords'].apply(converted)
```

```python
movies['keywords'][0]
```

```python
movies['cast']=movies['cast'].apply(converts)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  if isinstance(obj, str):  # Check if obj is a string before using ast.literal_eval
    l = []
    for i in ast.literal_eval(obj):
      if i['job'] == 'Director':
        l.append(i['name'])
        break
    return l
  else:  # If obj is not a string (e.g., a list), likely already processed
    # Check if it's a list and contains the director's name
    if isinstance(obj, list) and len(obj) > 0 and isinstance(obj[0], dict):
        for item in obj:
          if item.get('job') == 'Director':
              return [item.get('name')]
    return obj # Return as is if it doesn't have the expected format
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview'][0]
```

```python
#it is a string and we will convert it to list for concatination with others
# Check if the element is not already a list before splitting
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else x)
```

```python
# Remove spaces within column values to prevent inconsistencies in the recommender system
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#merging the columns to create a tag and converting them to string again to create a detailed paragraph for the mvoie tag
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tags'][0]
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#stemming the data to reduce words ex ['loving','love'] to ['love','love']
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Now our work is to recommend 5 similar movies to the user according to the movie or content he's watching
#To identitfy similarity we have to calculate the similarity between the tags of the movies
#therefore we will vectorise the text for that purpose and the techniques used is bag of words which is we will combine all the tags of all the mvoies
#then we will find the top 5000 words that has the most frequency from the concatenated tags
#for all the movies we will compare the rows that will have the frequencies of the top 5000 words for particular movies
#using CountVectorizer library for converting strings to token
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#toarray is used to convert the sparse matrix(from transform) to array
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
#gives us the most frequent words
cv.get_feature_names_out()
```

```python
#calculating cosine distances for finding the similarity between movies
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df[new_df['title']=='Avatar'].index
```

```python
#through enumerate function we could retain the original indexes of the movies even when they are reversed
#reversing the array to find the mostly similary movies on the top and then fetching top 5 similar movies from it
#using lambda function to check for the second value of tuple x (index,distance)
#1:6 is used to extarct top similar movies starting from 1st position as 0th position will have value 1 the highest as the movie have the distance with itself
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  #find the indexes of the movies
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
